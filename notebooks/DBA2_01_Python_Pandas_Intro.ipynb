# Basic Python commands
This course is not a full introduction to the python prgramming language. Instead we will use pandas and its commands to work with data. However there are some basics that you need to know. In the folloiwing example I am defining a function with two required and one optional argument. Note that you do not need to remember how to define functions. But you will need to understand the difference between required and optional arguments.

```python
def deduct(a, b, echo=False):
    result = a - b
    if echo:
        print(f'{a} - {b} = {result}')
    return result
```

## Passing Postional Arguments
I can call this function with two arguments, without defining the optional argument echo. Note that the order matters here: First we put in a and then b)

```python
deduct(5, 1)
```

If we change the order we get a different result

```python
deduct(1, 5)
```

## Setting Arguments in any order by using their name
We can also set required arguments with their name. This is optional but an improve readability. Its also allows us to put in arguments in any order. The result here is 4 because the finction calculates a-b.

```python
deduct(b=1, a=5)
```

## Optional Arguments
Optional arguments are different. They need to be called by their name. In the code above we did not set the optional argument "echo". This means the function was executed with the dault value, defined in the head of the function (echo=False). To rty this we can now call our function were we define the optional argument and set it to true. We can see that the function now prints its calulation.

```python
deduct(5,1,echo=True)
```

## Datatype Quick Guide
There are many different data types in python. Every depency/libary may even add more types. In order to work with pandas in this course we just need to know the following:
1) String: A sequence of characters
2) int: An interger number (a number without a fraction)
3) float: A number with a fraction
4) list: A list of values, each value can be of any type
5) dict: A dictionary which is a collection of key, value pairs
6) set: A list of values with the restriction that each value only occurs once
7) Pandas Dataframe: An object that stores data in a table-like format. Will be discussed in a later section.
8) Pandas Series: A Series of values all of the same datatype. Series are used to represent columns of dataframes.

### Datatype 5 dict - A dictionary which is a collection of key-value-pairs
The concept of stroing a value with a key is ingenious because it allows us to represent a "this for that" - relationship. In this course we will use it to replace or rename things.

```python
mydict = {
    "key1": "value1",
    "key2": "value2",
    "key3": "value3"
    }
type(mydict)
```

In contrast to lists, the order of the key-value pairs does not matter. We simply get our values by passing the repective key to the dictionary.

```python
mydict["key1"]
```

You can see how this can be usefull if we want to rename things. We could set the key to the value we want to replace and the value to what we want to replace with (this-for-that)

```python
rename = {
    "Kunde": "customer",
    "Produkt": "product",
    "Preis": "price"
    }
```

```python
rename["Produkt"]
```

### Datatype 6  set - A list of values with the restriction that each value only occurs once
While we may use sets as often because we do not need them so often since pandas has inbuilt functions to implent their functionality, I want to cover here to just be complete in the very basic datatypes of Python. Sets are like Lits with two mayor differences:
1) The order of elements does not matter
2) Every value is stored only once.

I will show the most common use case here: Finding unqiue values. For that we create a list which contains duplicate values. My truning that list into a set, the duplicates are removed.

```python
dup_list = ["Hans", "Peter", "Hans", "Peter", "Silvia", "Silvia", "Wili", "Peter", "Emma"]
my_set = set(dup_list)
my_set
```

```python
type(my_set)
```

Sets also support set operations, which are operations like. Which element is in A and B. What remains in A if i remove all of B and so on.

```python
my_set - set(["Hans", "Peter"])
```

### Which Series/Columns does my Dataframe have?
But which columns do we actually have in our dataframe? There are 3 ways to find this out
1) We can call the attribute _columns_ of the dataframe which will just give us a list (in Index Format)
2) We can call the attribute _dytes_ which will also give us all columns and the dataype each column has.
3) We can call the function _info()_ which give us the same as dtypes but allow shows information about missing values and the number of rows.

```python
tickets.columns
```

```python
tickets.dtypes
```

```python
tickets.info()
```
